# Sheriff and Bank Robber Problem

---

## Problem Statement

```python
#Sheriff has 180 km/h car 
#Bank robber has 150 km/h car and five-minute head start
#How long does it take the sheriff to catch the robber? 
#What distance will they have traveled at that point? 
#(For simplicity, letâ€™s ignore acceleration, traffic, etc.)
#Conversions: 150km/h -> 2.5 km/min. 180km/h -> 3 km/min
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,40,1000) # start, finish, n points
```

distance trabveled by robber d = 2.5t

```python
d_r = 2.5 * t
```

distance traveled by sheriff: d = 3(t-5)

```python
d_s = 3 * (t-5)
```

```python
fig, ax = plt.subplots()
plt.title('A Bank Robber Caught')
plt.xlabel('time (in minutes)')
plt.ylabel('distance (in km)')
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t, d_r, c='green')
ax.plot(t, d_s, c='brown')
plt.axvline(x=30, color='purple', linestyle='--')
plt.axhline(y=75, color='purple', linestyle='--')
plt.show()
```
